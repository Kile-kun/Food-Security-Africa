```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/Kile-kun/Food-Security-Africa/main/FoodBalanceSheets_E_Africa_NOFLAG.csv'
df_raw = pd.read_csv(url, encoding='latin-1')
df_raw.head()
```

```python

#11. What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

df_raw.groupby(['Item'])['Y2014', 'Y2017'].sum()
```

```python
#12. What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

df_raw.describe()
```

```python
#13. What is the total number and percentage of missing data in 2016 to 2 decimal places?
df_raw['Y2016'].isnull().sum()
```

```python
round((df_raw['Y2016'].isnull().sum()/df_raw['Y2016'].count())*100 ,2)
```

```python
# 14. Which year had the highest correlation with ‘Element Code’? 

df_raw.groupby(['Element Code'])['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018'].sum()
```

```python
#15. What year has the highest sum of Import Quantity?

df_raw.groupby(['Element'])['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018'].sum()
```

```python
#16. What is the total number of the sum of Production in 2014?

df_raw.groupby(['Element'])['Y2014'].sum()
```

```python
#17. Which of these elements had the highest sum in 2018?

df_raw.groupby(['Element'])['Y2018'].sum().rank(ascending=False)
```

```python
#18. Which of these elements had the 3rd lowest sum in 2018?

df_raw.groupby(['Element'])['Y2018'].sum().rank(ascending=True)
```
